#**Web Scraping - Unicorn Companies**#
data source: https://www.cbinsights.com/research-unicorn-companies



#Import necessary libraries

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_url = 'https://www.cbinsights.com/research-unicorn-companies/'
page = requests.get(data_url)
content = BeautifulSoup(page.text, 'html')
```

```python
content
```

```python
content.find('table', class_='wikitable sortable sticky-header')
```

```python
table = content.find_all('table')
print(table)
```

```python
titles = content.find_all('th')
```

```python
table_titles = [title.text.strip() for title in titles]

print(table_titles)
```

```python
#create an empty DataFrame to contain data
df = pd.DataFrame(columns = table_titles)

df
```

```python
column_data = content.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    company_data = [data.text.strip() for data in row_data]

    length = len(df)
    df.loc[length] = company_data
```

```python
df
```

```python
#export the dataframe to a csv file
df.to_csv(r'../content/Unicorn_company.csv', index = False)
```
